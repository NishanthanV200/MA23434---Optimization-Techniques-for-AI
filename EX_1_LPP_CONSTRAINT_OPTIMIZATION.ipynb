# **EX:1 LINEAR PROGRAMMING PROBLEM- CONSTRAINT OPTIMIZATION**

# Problem 1



---


maximize: z = 3x1 + 2x2



---



2x1 + x2 <= 20

---


4x1 - 5x2 <= 10

---


x1, x2 >= 0

---



```python
import numpy as np
from scipy.optimize import linprog
```

```python
c=[-3,-2]
```

```python
A=[
    [2,1],
    [4,-5]
]
```

```python
b=[20,10]
```

```python
x_bounds=(0,None)
y_bounds=(0,None)
```

```python
result=linprog(c,A_ub=A,b_ub=b,bounds=[x_bounds,y_bounds],method='highs')
```

```python
print("Optimal value:",-result.fun)
print("Values of x:",result.x)
```
